### 1) Migrant Families Survey Chhatarpur (DMKS - Darshna Mahila Kalyan Samiti)

```python
Image("overview.png", width = 600, height = 150) # from Panchayat Survey Report
```

```python
df_C = pd.read_excel(path_raw / 'DATA_ANALYSIS Migrant Family Survey Chhatarpur-3676360759(DMKS 08.03.21).xlsx')
```

```python
df_C.head()
```

### Total number of respondents

```python
total_no = len(df_C['Identifier'])
total_no
```

#### What safety practices against COVID-19 are you aware of?

```python
mask_use = sum(df_C['--OPTION--|Use of Masks'])
print(f"Use of Masks: {mask_use}, {mask_use / total_no *100 :.2f}%")
```

```python
handwash_soap = sum(df_C['--OPTION--|Handwashing using soap'])
print(f"Handwashing using soap: {handwash_soap}, {handwash_soap / total_no *100 :.2f}%")
```

```python
social_dist = sum(df_C['--OPTION--|Social Distancing'])
print(f"Social Distancing: {social_dist}, {social_dist / total_no *100 :.2f}%")
```

```python
symptoms = sum(df_C['--OPTION--|Awareness of symptoms'])
print(f"Awareness of symptoms: {symptoms}, {symptoms / total_no *100 :.2f}%")
```

#### Do you have COVID-19 helpline number

```python
yes_nbr = df_C[df_C['Do you have Coronavirus/COVID-19 helpline number?'] == 'Yes']['Identifier'].count()
print(f"Yes = {yes_nbr}, {yes_nbr / total_no *100 :.2f}%")
```

```python
no_nbr = df_C[df_C['Do you have Coronavirus/COVID-19 helpline number?'] == 'No']['Identifier'].count()
print(f"No = {no_nbr}, {no_nbr / total_no *100 :.2f}%")
```

#### Do you know where is COVID-19 testing / treatment centre in your area?

```python
yes_test = df_C[df_C['Do you know where is COVID-19 testing / treatment centre in your area?'] == 'Yes']['Identifier'].count()
print(f"Yes = {yes_test}, {yes_test / total_no *100 :.2f}%")
```

```python
no_test = df_C[df_C['Do you know where is COVID-19 testing / treatment centre in your area?'] == 'No']['Identifier'].count()
print(f"No = {no_test}, {no_test / total_no *100 :.2f}%")
```

#### Were your family members tested for COVID-19 after coming back from migration?

```python
yes_covtest = df_C[df_C['Were your family members tested for COVID-19 after coming back from migration?'] == 'Yes']['Identifier'].count()
print(f"Yes = {yes_covtest}, {yes_covtest / total_no *100 :.2f}%")
```

```python
no_covtest = df_C[df_C['Were your family members tested for COVID-19 after coming back from migration?'] == 'No']['Identifier'].count()
print(f"No = {no_covtest}, {no_covtest / total_no *100 :.2f}%")
```

#### For how much time do you have sufficient food items in your house?

```python
food_1d = df_C[df_C['For how much time do you have sufficient food items in your house?'] == 'For 1-2 Days']['Identifier'].count()
print(f"For 1-2 Days = {food_1d}, {food_1d / total_no *100 :.2f}%")
```

```python
food_2d = df_C[df_C['For how much time do you have sufficient food items in your house?'] == '2-7 Days']['Identifier'].count()
print(f"2-7 Days = {food_2d}, {food_2d / total_no *100 :.2f}%")
```

```python
food_15d = df_C[df_C['For how much time do you have sufficient food items in your house?'] == '15 days to 1 month']['Identifier'].count()
print(f"15 days to 1 month = {food_15d}, {food_15d / total_no *100 :.2f}%")
```

```python
food_2m = df_C[df_C['For how much time do you have sufficient food items in your house?'] == 'For more than 2 months']['Identifier'].count()
print(f"For more than 2 months = {food_2m}, {food_2m / total_no *100 :.2f}%")
```
